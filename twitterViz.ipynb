```python
'''
This cell loads the data from the .json file and stores it into:
 -- a list of dictionaries called tweetsList
 -- a pandas DataFrame called tweetsDF

You can assume that this cell will be executed before any of the visualization cells.

The cell includes the necessary imports for working with MatplotLib, and Seaborn
(as well as pandas and JSON files), but you can add others here as needed.

This cell will also display for use as a reference:
 -- the first tweet from tweetsList
 -- the first five tweets from tweetsDF
 
(This cell will also take a little while to run...)
'''

import json
import matplotlib.pyplot as plt
import seaborn as sns
import pprint
import pandas as pd

sns.set()

# Load dataset from the .json tweets file
with open('covidTweets.json', 'r', encoding="utf-8") as infile:
    tweetsList = json.load(infile)

# Convert the dataset into a partially-flattened pandas DataFrame   
tweetsDF = pd.io.json.json_normalize(tweetsList, sep='_')

# Sample output
print("The first tweet from tweetsList is:\n")
pprint.pprint(tweetsList[0])    

print("\n\nThe first five tweets from tweetsDF are:")
tweetsDF.head()
```

Visualization Question 1: What is the relationship between a status verified user vs. a standard user?

Chart: Scatterplot

User Learned: 
Based on the data, we can see that the more followers a user has, the less inclined they are to tweet publicly. Users who are not too popular tend to post frequently. This could be that a popular user feels 'scared' to post and speak their mind because of potential judgement by their mass following. This fear is mainly due to "cancel culture" in the Twitter world where celebrities often get cyber bullied if they aren't to careful of what they say. Because of this culture, brands are less willing to sponsor them and this also makes it hard for celebrities to rise back up again.  People with less followers don't have that many people to be judged by and therefore have more freedom to post their opinions.

```python
df1 = tweetsDF[['user_followers_count','user_statuses_count']] # surfs through the COVID tweets database
#print(df1)

followers_list = tweetsDF['user_followers_count'].values.tolist() # number of followers a user has converted to list
#print(retweet_list)

statuses_list = tweetsDF['user_statuses_count'].values.tolist() # number of tweets a user has tweeted
%matplotlib inline
plt.scatter(followers_list, statuses_list)
plt.title('Followers vs. Number of Tweets Comparison') # labels title of graph
plt.xlabel('Followers') # labels x-axis
plt.ylabel('# of Tweets') # labels y-axis
plt.show()
```

Visualization Question 2: What is the average number of followers per user?

Chart: Signal graph

User Learned: 
We can see that the graph peaks at 60,000 followers per user. We can determine that 60,000 followers is the average number of followers an active/popular twitter user has. Following that determination, most users also have a number range from 0 to 20,000 followers, and fewer user have above 60,000 followers. 

```python
plt.plot(followers_list) # get the signal graph
plt.title('Accumulation of User Followers')
plt.xlabel('Number of Followers')
plt.ylabel('Increase >>')
plt.show()
```

Visualization Question 3: What is the highest amount in distribution of the number of tweets a user posts?

Chart: Bar Chart/Histogram

User Learned:
According to the bar chart, we can see that the highest number of tweets a user has tweeted is at a cap of around 90,000 tweets. Learned that a user has not posted more than 100,000 tweets and a typical range of posts is between 0 to 90,000.

Visualization Question 4: How does tweets posted affect the relationship of how many people the user follows?

Chart: Line Plot

User Learned:
By looking at the plot, we can see that the more a user tweets, the less number of people the user follows. This logic makes sense given the COVID-19 tweets database. A user is more likely to get informed if they follow more and more people which leaves users to seek information rather than posting their own tweets. Users who follow don't follow that many people may be considered an influencer to other users which explains the higher ratio towards the number of tweets they post.

```python
following_list = tweetsDF['user_friends_count'].values.tolist() # convert the number of people a user follows into a list
plt.plot(statuses_list, following_list) # plot a line graph

plt.title('Distribution of Tweets')
plt.xlabel('Number of Tweets')
plt.ylabel('User Following')
plt.show()
```

Visualization Question 5: Does ratio of followers and/or following affect the number of times a user tweets?

Chart: Double Scatterplot

User Learned:
By comparing three sets of data, we are able to tell that people who follow more people and have a lesser amount of followers tend to tweet more than people with more followers than following. This logic can help when finding which users are reliable in their tweets when looking at people with more followers. This finding also helps with factoring out unpopular and popular user tweets.

```python
plt.plot(following_list, statuses_list, 'go', label='Following') # creates legend for first set of plot
plt.plot(followers_list, statuses_list, 'y*', label='Followers') # second set
plt.title('Followers vs. Following on the Number of Tweets')
plt.xlabel('Followers/Following')
plt.ylabel('Number of Tweets')
plt.legend(loc='best')  # legend text
plt.show()
```
